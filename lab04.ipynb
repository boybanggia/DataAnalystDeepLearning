
# **Plotting with categorical data**

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

# Biểu đồ phân tán Categorical 


```python
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips)
```

**Các jitterthông số kiểm soát độ lớn của jitter hoặc vô hiệu hóa nó hoàn toàn:**

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

# **Phân bố các quan sát trong các danh mục**

**Boxplots**

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
#Khi thêm huengữ nghĩa, hộp cho mỗi cấp của biến ngữ nghĩa được di chuyển dọc theo trục phân loại để chúng không trùng lặp:
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
```

```python
#Hành vi này được gọi là "dodging" và được bật theo mặc định vì nó được giả định rằng biến ngữ nghĩa được lồng trong biến phân loại chính. Nếu không phải như vậy, bạn có thể tắt tính năng né tránh:
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.catplot(x="day", y="total_bill", hue="weekend",
            kind="box", dodge=False, data=tips)
```

Một chức năng có liên quan **boxenplot**


```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(x="color", y="price", kind="boxen",
            data=diamonds.sort_values("color"))
```

# **Ước tính thống kê trong các danh mục**

```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
```

```python
sns.catplot(y="deck", hue="class", kind="count",
            palette="pastel", edgecolor=".6",
            data=titanic)
```

# **Biểu đồ điểm** 

```python
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
```

```python
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic)
```

# **Lập đồ thị dữ liệu “dạng rộng”**

```python
iris = sns.load_dataset("iris")
sns.catplot(data=iris, orient="h", kind="box")
```

```python
sns.violinplot(x=iris.species, y=iris.sepal_length)
```

# **Hiển thị nhiều mối quan hệ với các khía cạnh**

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.7,
            kind="swarm", data=tips)
```

```python
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log")
```
